La altura de una población sigue una distribución normal con media
μ = 160 cm y desviación estándar σ = 10 cm.
¿Cuál es la probabilidad de que una persona seleccionada al azar
tenga una altura mayor a 170 cm?

```python
# Importar la biblioteca SciPy
import scipy.stats as stats

# Definir la media y la desviación estándar
media = 160
desviacion = 10

# Calcular la probabilidad de que x sea mayor a 170
probabilidad = 1 - stats.norm.cdf(170, loc=media, scale=desviacion)

print("La probabilidad de que x sea mayor al 170:", probabilidad)
```

¿Y cuál es la probabilidad de que la altura sea menor o igual a 155
cm?

```python
# Importar la biblioteca SciPy
import scipy.stats as stats

# Definir la media y la desviación estándar
media = 160
desviacion = 10

# Calcular la probabilidad de que x sea menor o igual a 155
probabilidad = stats.norm.cdf(155, loc=media, scale=desviacion)

print("La probabilidad de que x sea menor o igual a 155:", probabilidad)
```

Encuentra la probabilidad de que una persona tenga una altura entre
150 cm y 165 cm.

```python
import scipy.stats as stats

# Definir la media y la desviación estándar
media = 160
desviacion = 10

# Calcular la probabilidad de que x esté entre 150 y 160
probabilidad = stats.norm.cdf(110, loc=media, scale=desviacion) - stats.norm.cdf(90, loc=media, scale=desviacion)

print("La probabilidad de que x esté entre 150 y 160 es:", probabilidad)
```

Si el 20 % de las personas más altas son consideradas muy altas, ¿cuál
es la altura a partir de la cual una persona puede ser considerada muy
alta?

```python
import scipy.stats as stats

# Parámetros de la distribución
media = 160
desviacion = 10

# Calcular la altura mínima para el 20% más alto (percentil 80)
altura_muy_alta = stats.norm.ppf(0.80, loc=media, scale=desviacion)

print("Altura mínima para ser considerado muy alto:", altura_muy_alta)
```
