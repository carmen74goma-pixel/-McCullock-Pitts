Predecir el PIB de un país según exportaciones e inversión.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# --- 1. Preparación de Datos y Ajuste del Modelo ---

# Datos (en miles de millones)
Exportaciones = np.array([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])
Inversiones = np.array([50, 100, 150, 200, 250, 300, 350, 400, 450, 500])
PIB = np.array([600, 1200, 1800, 2500, 3100, 3700, 4200, 4800, 5300, 5900])

# Matriz de características X (Exportaciones e Inversiones)
X = np.column_stack((Exportaciones, Inversiones))
y = PIB

# Ajustar modelo de Regresión Lineal Múltiple
model = LinearRegression()
model.fit(X, y)

# Coeficientes
beta_0 = model.intercept_
beta_1 = model.coef_[0]  # Coeficiente para Exportaciones
beta_2 = model.coef_[1]  # Coeficiente para Inversiones

print(f"--- Coeficientes del Modelo ---")
print(f"Ecuación del modelo: PIB = {beta_0:.2f} + {beta_1:.2f} * Exportaciones + {beta_2:.2f} * Inversiones")
print(f"Coeficiente para Exportaciones (beta_1): {beta_1:.2f}")
print(f"Coeficiente para Inversiones (beta_2): {beta_2:.2f}")
print("---------------------------------")


# --- 2. Respuesta a Pregunta 1: Predicción ---
# Exportaciones = 1200, Inversiones = 600 (en miles de millones)
X_new_Q1 = np.array([[1200, 600]])
PIB_estimado_Q1 = model.predict(X_new_Q1)[0]
print(f"PIB estimado para Exportaciones=1200 y Inversiones=600: {PIB_estimado_Q1:.2f} mil millones")
print("---------------------------------")


# --- 3. Respuesta a Pregunta 2: Impacto de aumento del 10% en Exportaciones ---
# Se usará un caso base representativo (el último punto de datos: Exp=1000, Inv=500)
# Aumento del 10% en Exportaciones: de 1000 a 1100 (un cambio de 100 mil millones)
delta_Exportaciones = 100 # mil millones
delta_PIB_abs = beta_1 * delta_Exportaciones

# PIB base predicho (para calcular el cambio porcentual)
X_base = np.array([[1000, 500]])
PIB_base_pred = model.predict(X_base)[0]

delta_PIB_percent = (delta_PIB_abs / PIB_base_pred) * 100
print(f"Si las Exportaciones aumentan en 10% (i.e., {delta_Exportaciones} mil millones) y la Inversión se mantiene constante, el PIB aumentará en {delta_PIB_abs:.2f} mil millones.")
print(f"Esto representa un aumento porcentual del PIB de aproximadamente {delta_PIB_percent:.2f}% (basado en el caso de Exp=1000, Inv=500).")
print("---------------------------------")


# --- 4. Respuesta a Pregunta 3: Impacto de aumento de 1 millón en Inversión ---
# Cambio en Inversión = 1 millón = 0.001 mil millones
delta_Inversion_millions = 1
delta_Inversion_billions = 0.001
delta_PIB_Q3 = beta_2 * delta_Inversion_billions
print(f"Impacto de un aumento de {delta_Inversion_millions} millón en la Inversión sobre el PIB: {delta_PIB_Q3:.4f} mil millones")
print("---------------------------------")


# --- 5. Respuesta a Pregunta 4: Combinación para PIB Objetivo de 7000 ---
# La ecuación es: Exportaciones * beta_1 + Inversiones * beta_2 = 7000 - beta_0
PIB_objetivo = 7000
valor_derecha = PIB_objetivo - beta_0

# Proporcionamos una combinación específica asumiendo Inversiones = 600 mil millones
I_supuesta = 600
E_necesaria = (valor_derecha - (I_supuesta * beta_2)) / beta_1
print(f"Para un PIB objetivo de {PIB_objetivo} mil millones, la relación es:")
print(f"{beta_1:.2f} * Exportaciones + {beta_2:.2f} * Inversiones = {valor_derecha:.2f}")
print(f"Una combinación posible es: Exportaciones = {E_necesaria:.2f} mil millones e Inversiones = {I_supuesta:.2f} mil millones.")
print("---------------------------------")


# --- 6. Respuesta a Pregunta 5: Variabilidad Explicada (R^2) ---
y_pred_train = model.predict(X)
r_squared = r2_score(y, y_pred_train)
print(f"R-cuadrado (R^2) del modelo: {r_squared:.4f}")
print(f"El modelo explica el {r_squared * 100:.2f}% de la variabilidad del PIB.")
print("---------------------------------")
```
